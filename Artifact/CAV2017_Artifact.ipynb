We start by comparing K-Nearest Neighbors and the Lexicographic projection for STL

K-Nearest Neighbors implementation: https://github.com/slaypni/fastdtw

## Define DTW distance measure function

```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

def dtw_dist(x, y):
    return fastdtw(x['Y'].sample(0.1), y['Y'].sample(0.1), dist=euclidean)[0]
```

## Define STL distance measure function
Want: 
$$
\phi = \square\left(\text{step}(x_{ref}) \implies \Diamond_{[0, \tau]}(x - x_{ref} > a)\right)
$$

If we take $step(x_{ref})$ to be $\dot{x}_{ref} > 5$, then for our set of traces $\phi$ simplifies to:

$$\phi = F_{[10, \tau]}(x - x_{ref}) > a$$

```python
phi = stl.parse("(F[10.15, tau?](Y > a?))")
```

```python
def dist(x, y):
    return np.linalg.norm(x - y)


def val_to_vec(val):
    return np.array([x for _, x in sorted(list(val.items()))])

def project(x):
    order = ("a?", "tau?")
    ranges = {"a?": (-3, 3), "tau?": (10.0001, 1000)}
    polarity = {"a?": False, "tau?": True} # Monotonically Increasing or Decreasing
    return stl.lex_param_project(
        phi, x, order=order, ranges=ranges, polarity=polarity)

DISTS = np.vstack(list(map(lambda x: val_to_vec(project(x)), traces)))
stl_sigma = np.var(DISTS, axis=0)
stl_mu = np.mean(DISTS, axis=0)

def stl_dist(x, y):
    return dist(*map(lambda a: (val_to_vec(project(a)) - stl_mu)/stl_sigma, (x, y)))
```

## Compute the Adjacency Matricies

```python
from itertools import product

M_dtw = np.zeros((len(traces), len(traces)))
M_stl = np.zeros_like(M_dtw)
for (i, x), (j, y) in product(enumerate(traces), enumerate(traces)):
    M_dtw[i, j] = dtw_dist(x, y)
    M_stl[i, j] = stl_dist(x, y)

A_dtw = 1 - M_dtw/np.max(M_dtw) - np.eye(7)
A_stl = 1 - M_stl/np.max(M_stl) - np.eye(7)
```

```python
def plot_adjacency(A):
    plt.pcolor(A, cmap="Blues")
    plt.xticks(np.arange(0.5, 7), range(7))
    plt.yticks(np.arange(0.5, 7), range(7))
    plt.xlabel("Trace")
    plt.ylabel("Trace")
    plt.colorbar()
    plt.xticks

plot_adjacency(A_dtw)
plt.show()
```

```python
plot_adjacency(A_stl)
plt.show()
```

```python
from sklearn.cluster import SpectralClustering
C = SpectralClustering(n_clusters=3, affinity="precomputed")
l_dtw = C.fit_predict(A_dtw)
C = SpectralClustering(n_clusters=3, affinity="precomputed")
l_stl = C.fit_predict(A_stl)
```

```python
from collections import Counter

def set_index(df):
    # reset time
    df.index -= df.index[0]
    # Time per frame: 0.1 sec
    df.index *= 0.1
    return df

def is_not_trival(x):
    lane_ids = x["Lane_ID"]
    seen_lanes = {v for _, v in lane_ids.items()}
    return ({2,3,4} & seen_lanes) 

def load_signals(paths, drop_trivial=True):
    signals = (from_pandas(set_index(pd.read_csv(path, index_col="Frame_ID", 
            usecols=["Frame_ID", "Lane_ID", "Vehicle_ID"]))) 
            for path in paths)
    if drop_trivial:
        signals = filter(is_not_trival, signals)
    return signals
```

### Specs

$$
\psi_i \triangleq \Diamond\bigg[(x \notin L_i) \wedge \left(\Diamond_{(0, \epsilon]}(x \in L_i\ U_{[\epsilon, \tau_i]}\ x \notin L_i) \right) \bigg ]$$

$\epsilon = 0.5$

```python
lanes = list(range(2, 5))
in_lane_aps = {stl.parse("in_lane_{}".format(i)): stl.parse("Lane_ID = {}".format(i)) for i in lanes}
```

```python
psi_template = "F((in_lane_{i}) & (F[0, 0.5]((in_lane_{i}) U[0.5, tau?] (~(in_lane_{i})))))"
psis = {i: stl.parse(psi_template.format(i=i)) for i in lanes}
psis = {i: stl.utils.inline_context(psi, in_lane_aps) for i, psi in psis.items()}
```

```python
def project(x):
    order = ["tau?"]
    ranges = {"tau?": (2, 10000)}
    polarity = {"tau?": True}
    return [stl.lex_param_project(
        psis[i], x, order=order, ranges=ranges, polarity=polarity)['tau?'] for i in lanes]
```

```python
def project_signals(paths):
    for signal in load_signals(paths):
        try:
            yield project(signal)
        except:
            pass
```

## Reorient

$$\square_{[0, \tau]}(y \leq y_{min} \wedge x \leq x_{max})$$

```python
phi = stl.parse("G[0, tau?]((y <= yMin?) and (x <= xMax?))")

def project(x):
    order = ("tau?", "yMin?", "xMax?")
    # Use tau Domain from paper
    ranges = {"yMin?": (-7, 15), "xMax?": (-7, 15), "tau?": (10, 60)}
    polarity = {"xMax?": True, "tau?": False, "yMin?": True}
    return stl.lex_param_project(phi, x,order=order, polarity=polarity, 
                                 ranges=ranges, tol=0.5)
```
